```python
def metric_four_plot(x1, x2, x3, x4, y1, y2, y3, y4, metric='MAE'):
    plt.xlabel('Prediction Step')
    plt.plot(x1, y1, 'b-', label=metric+'(LSTM)')
    plt.plot(x2, y2, 'g-.', label=metric+'(K=1)')
    plt.plot(x3, y3, 'r--', label=metric+'(K=12)')
    plt.plot(x4, y4, label=metric+'(Transformer)')
    plt.legend()
    if metric is 'MAPE':
        for i in range(len(x1)):
            plt.text(x1[i], y1[i], format(y1[i], '.2%'))
        for i in range(len(x2)):
            plt.text(x2[i], y2[i], format(y2[i], '.2%'))
        for i in range(len(x3)):
            plt.text(x3[i], y3[i], format(y3[i], '.2%'))
        for i in range(len(x4)):
            plt.text(x4[i], y4[i], format(y4[i], '.2%'))
    else:
        for i in range(len(x1)):
            plt.text(x1[i], y1[i], format(y1[i], '.2f'))
        for i in range(len(x2)):
            plt.text(x2[i], y2[i], format(y2[i], '.2f'))
        for i in range(len(x3)):
            plt.text(x3[i], y3[i], format(y3[i], '.2f'))
        for i in range(len(x4)):
            plt.text(x4[i], y4[i], format(y4[i], '.2f'))
```

```python
def metric_double_plot(x1, x2, y1, y2, metric='MAE'):
    plt.xlabel('Prediction Step')
    plt.plot(x1, y1, 'b-', label=metric+'(LSTM)')
    plt.plot(x2, y2, 'r--', label=metric+'(K=12)')
    plt.legend()
    if metric is 'MAPE':
        for i in range(len(x1)):
            plt.text(x1[i], y1[i], format(y1[i], '.2%'))
        for i in range(len(x2)):
            plt.text(x2[i], y2[i], format(y2[i], '.2%'))
    else:
        for i in range(len(x1)):
            plt.text(x1[i], y1[i], format(y1[i], '.2f'))
        for i in range(len(x2)):
            plt.text(x2[i], y2[i], format(y2[i], '.2f'))
```
